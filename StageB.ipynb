```python
import pandas as pd
import numpy as np
from sklearn import preprocessing,model_selection, metrics

data = pd.read_csv("energydata_complete.csv")
data.head()
```

# removing the following columns: [“date”, “lights”]. 

```python
data = data.drop(['date', 'lights'], axis=1)
data.head()
```

# using the MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

data_scaled = pd.DataFrame(scaler.fit_transform(data), columns = data.columns)
```

# The target variable is “Appliances”

```python
x = data_scaled.drop(columns=['Appliances'])
y = data_scaled['Appliances']
```

# Use a 70-30 train-test set split with a random state of 42 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=42)
```

# From the dataset, fit a linear model on the relationship between the temperature in the living room in Celsius (x = T2) and the temperature outside the building (y = T6). What is the R^2 value in two D.P

```python
X2 = data_scaled['T2'].values.reshape(-1, 1)
Y2 = data_scaled['T6'].values.reshape(-1, 1)

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X2, Y2, test_size = 0.3, random_state=42)

from sklearn import linear_model 

lin_regr2 = linear_model.LinearRegression()
lin_regr2.fit(xtrain, ytrain)

pred2 = lin_regr2.predict(xtest)

from sklearn.metrics import r2_score
r2_score  = r2_score(ytest, pred2) 
round(r2_score, 2) 
```

# What is the Mean Absolute Error (in two decimal places)for the main dataset?

```python
from sklearn import linear_model 
lin_regr = linear_model.LinearRegression()
lin_regr.fit(x_train, y_train)

pred = lin_regr.predict(x_test)

from sklearn.metrics import mean_absolute_error
MAE = mean_absolute_error(y_test, pred)
round(MAE, 2)
```

# What is the Residual Sum of Squares (in two decimal places) for the main dataset?

```python
RSS = np.sum(np.square(y_test - pred))
round(RSS, 2)
```

# What is the Root Mean Squared Error (in three decimal places) for the main dataset?

```python
from sklearn.metrics import mean_squared_error
RMSE = np.sqrt(mean_squared_error(y_test, pred))
round(RMSE, 3)
```

# What is the Coefficient of Determination (in two decimal places)?

```python
#R-Squared or coefficient of determination
from sklearn.metrics import r2_score

r2_score = r2_score(y_test, pred)
round(r2_score, 2)
```

# Obtain the feature weights from your linear model above. Which features have the lowest and highest weights respectively?

```python
def get_weights(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    
    return weights_df

linear_model_weights = get_weights(lin_regr, x_train, 'Weight')
linear_model_weights 
```

# observation : RH_2 has the lowest weights while RH_1 has the highest weights


# Train a ridge regression model with an alpha value of 0.4. Is there any change to the root mean squared error (RMSE) when evaluated on the test set?

```python
from sklearn.linear_model import Ridge 

ridge = Ridge(alpha=0.4)
ridge.fit(x_train, y_train)
pred_ridge = ridge.predict(x_test)

from sklearn.metrics import mean_squared_error

RMSE = np.sqrt(mean_squared_error(y_test, pred_ridge))
round(RMSE, 3)
```

# There is no change in RMSE 

# Train a lasso regression model with an alpha value of 0.001 and obtain the new feature weights with it. How many of the features have non-zero feature weights?

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.001)
lasso.fit(x_train, y_train)

pred_lasso = lasso.predict(x_test)

#Lasso on weights
lasso_weights = get_weights(lasso, x_train, 'Lasso-weight')
lasso_weights 
```

#Four features have non-zero weights. They include RH_out, RH_8, Windspeed, RH_1

# What is the new RMSE with the Lasso Regression (in 3 decimal places)?

```python
from sklearn.metrics import mean_squared_error

RMSE = np.sqrt(mean_squared_error(y_test, pred_lasso))
round(RMSE, 3)
```
